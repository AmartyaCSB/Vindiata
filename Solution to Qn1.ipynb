```python
# Importing Packages
import numpy as np
import pandas as pd
```

```python
# Importing and Reading Files using Pandas
xls = pd.ExcelFile('Case_study.xlsx')

# Operations Sheet in a Dataframe
df1 = pd.read_excel(xls, 'Operations')

# AC Characeristics in a Dataframe
df2 = pd.read_excel(xls, 'AC characteristics')
```

```python
# Grouping of Aircraft type to calculate Total Hours flown by Airline A
df1 = df1.groupby(['Aircraft Type']).sum() 
df1['Total hours'] = df1.sum(axis=1)
```

```python
# Merging and cleaning Data (Done to keep it clean,though unnecessary)
df3 = pd.merge(df2, df1, left_on='Aircraft Type', right_index=True, how='left', sort=False);
df = df3.drop(['Jan-14','Feb-14','Mar-14','Apr-14','May-14','Jun-14','Jul-14','Aug-14','Sep-14','Oct-14','Nov-14','Dec-14'], axis=1)
```

```python
# Functions to carry out the Calculations
def Product(x, y):
    return x * y 
def Division(x,y):
    return (x/y)
```

```python
# 'Total cost' per Aircraft Carrier calculated 

df['Total Cost'] = df.apply(lambda row: Product(row['Costs per flight hour'], row['Total hours']), axis=1)
df
```

```python
# Calculations done to find out the Cost Per Seat per Distance flown for each aircraft by Airline A

df['Dist Flown'] = df.apply(lambda row: Product(row['Ave. Speed (km/h)'], row['Total hours']), axis=1)
df['Cost per Dist'] = df.apply(lambda row: Division(row['Total Cost'], row['Dist Flown']), axis=1)
df['Cost per Dist per Seat'] = df.apply(lambda row: Division(row['Cost per Dist'], row['Number of Seats']), axis=1)
df
```
